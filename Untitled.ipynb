```python
nltk.download('punkt')
```

```python
#messages tokenization
from nltk.tokenize import word_tokenize
tokenized_msg = [word_tokenize(msg) for msg in messages["content"]]
```

```python
tokenized_msg
```

```python
#punctuation removal
import re
regex = re.compile('[%s]' % re.escape(string.punctuation))

tokenized_msg_punctuation_free = []

for msg in tokenized_msg:
    new_msg = []
    for token in msg:
        new_token = regex.sub(u'', token)
        if not new_token == u'':
            new_msg.append(new_token)
    
    tokenized_msg_punctuation_free.append(new_msg)
```

```python
tokenized_msg_punctuation_free
```

```python
nltk.download('stopwords')
```

```python
#stopwords removal
from nltk.corpus import stopwords

tokenized_msg_stopwords_free = []

for msg in tokenized_msg_punctuation_free:
    new_term_vector = []
    for word in msg:
        if not word in stopwords.words('english'):
            new_term_vector.append(word)
    
    tokenized_msg_stopwords_free.append(new_term_vector)
```

```python
tokenized_msg_stopwords_free
```

```python
nltk.download('wordnet')
```

```python
# Stemming and Lemmatization
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

stemm = PorterStemmer()
lemm = WordNetLemmatizer()

preprocessed_msg = []

for msg in tokenized_msg_stopwords_free:
    final_msg = []
    for word in msg:
        #final_msg.append(stemm.stem(word))
        final_msg.append(lemm.lemmatize(word))
    
    preprocessed_msg.append(final_msg)
```

```python
preprocessed_msg
```

```python
import gensim
from gensim import corpora
```

```python
#lda model
# Create Dictionary
id2word = corpora.Dictionary(preprocessed_msg)

# Create Corpus: Term Document Frequency
#convert the dictionary to a bag of words
corpus = [id2word.doc2bow(text) for text in preprocessed_msg]

# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=5, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=10,
                                           passes=10,
                                           alpha='symmetric',
                                           iterations=100,
                                           per_word_topics=True)
```

```python
lda_model.print_topics()
```

```python
import logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
def send_message_to_mail_box(user, pwd, recipient, subject, body):
    import smtplib

    FROM = user
    TO = recipient if isinstance(recipient, list) else [recipient]
    SUBJECT = subject
    TEXT = body

    # Prepare actual message
    message = """From: %s\nTo: %s\nSubject: %s\n\n%s
    """ % (FROM, ", ".join(TO), SUBJECT, TEXT)
    try:
        server = smtplib.SMTP("smtp.gmail.com", 587)
        #server = smtplib.SMTP_SSL(host='',port=465)
        server.ehlo()
        server.starttls()
        server.login(user, pwd)
        server.sendmail(FROM, TO, message)
        server.close()
        print ('successfully sent the mail')
    except Exception as e:
        logger.error('failed to send mail \t'+str(e))
```
